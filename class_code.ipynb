# J(Theta) = theta^2 + 5*sin(theta)


```python
import numpy as np
import matplotlib.pyplot as plt

# define theta
theta = np.linspace(-5, 5, 1000)

# define J(theta)

def J(x):
    return 0.3*x**2 + np.sin(3*x)

# define dJ(theta) - derivative of J(theta)
def dJ(x):
    return 0.6*x + 3*np.cos(x)

# plot J(theta)
J_theta = J(theta)
plt.plot(theta, J_theta, 'b')
plt.grid(axis='both')
plt.show()



```

```python
theta0 = 2* np.random.randn(1)# initial guess
alpha = 0.001 # learning rate
J_theta0 = J(theta0) # initial cost
print("J(theta0) = ", J_theta0) # print initial cost
plt.plot(theta, J_theta, 'b' , theta0, J_theta0, 'ro') # plot J(theta) and initial guess
plt.grid(axis='both') # add grid
plt.show() # show plot
```

```python
# gradient descent
for i in range(1000):
    theta0 = theta0 - alpha * dJ(theta0) # update theta
    J_theta0 = J(theta0) # update cost
    if i % 100 == 0:
        plt.plot(theta, J_theta, 'b' , theta0, J_theta0, 'ro') # plot J(theta) and current guess
        plt.grid(axis='both') # add grid
        plt.show() # show plot
```
